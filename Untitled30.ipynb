```python
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import base64
import csv
import seaborn as sns
import numpy as np

#lectura de archivo csv
""" Lectura de archivo CSV separado por';'"""
datos = pd.read_csv('200511COVID19MEXICO.csv',sep=',',header=0,encoding='latin1')
datos=pd.DataFrame(datos)

#Modificacion para su uso y creacion de dataframe base

datos['FECHA_ACTUALIZACION']=pd.to_datetime( datos['FECHA_ACTUALIZACION'],dayfirst=True)
datos['FECHA_INGRESO']=pd.to_datetime( datos['FECHA_INGRESO'],dayfirst=True)
datos['FECHA_SINTOMAS']=pd.to_datetime( datos['FECHA_SINTOMAS'],dayfirst=True)
datos['FECHA_DEF']=datos['FECHA_DEF'].replace('9999-99-99','NaN')
datos['FECHA_DEF']=pd.to_datetime( datos['FECHA_DEF'],errors='ignore',dayfirst=True)
datos['DEFUNCIONES']=datos['FECHA_DEF']#se aderio esta columna al df original



########################################################## PARTE 1 ###########################################################
#filtrado de informacion
datos1=datos.loc[datos['RESULTADO'] ==1] # se filtro de acuerdo a pacientes que dieron positivo
datos1=datos.loc[datos['ENTIDAD_NAC'] ==26] # y en el estado de sonorA

#Dataframe 1
a=pd.DataFrame(datos1[['FECHA_SINTOMAS','RESULTADO']] #SE CREO EL PRIMER DF CON LA FECHA DE LOS SINTOMAS Y LOS RESULTADOS YA FILTRADOS
datosc=a.rename(columns={'FECHA_SINTOMAS':'FECHA','RESULTADO':'CONFIRMADOS'} )#se renombro la fecha
datosc=datosc.groupby((['FECHA']),as_index=False).count()#Se agrupo de acuerdo a la fecha para evitar repeticion de valores
                                                        #mayormente se uso count() en lugar de sum(), dado que cada registro
                                                        #corresponde a un solo paciente,  y asi evitamos confundir al programas
                                                           #con los valores registrados
               

#Dataframe 2
b= pd.DataFrame(datos1[['FECHA_DEF','DEFUNCIONES']]) #Dataframe 2-- Defunciones y Fecha defunciones
datosm=b.rename(columns={'FECHA_DEF':'FECHA','DEFUNCIONES':'DEFUNCIONES'} )#se renombro la fecha
datosm=datosm.groupby((['FECHA']),as_index=False).count()#agrupacion por fecha
#Concatenacion de dataframes
dat1=pd.concat([datosc,datosm],ignore_index= 'False',)#se concatena 
dat1=dat1.groupby((['FECHA']),as_index=False).sum()# se agrupa  y se evita qe la fecha llegue al indice
#Guardar como CSV
dat1.to_csv(r'C:\Users\manuel\tabla1.csv')# se exporta a csv


####################################################### PARTE 2 ##################################################################
#Dataframes filtradas
datosSn=datos.loc[datos['ENTIDAD_NAC'] ==26]#filtro por estado
datosSn=pd.DataFrame(datosSn)#df de estado
datosNl= datos.loc[datos['ENTIDAD_NAC'] ==19]#filtro por estado
datosNl=pd.DataFrame(datosNl)#df de estado
datosPb=datos.loc[datos['ENTIDAD_NAC']==21]#filtro por estado
datosPb=pd.DataFrame(datosPb)#df de estado
datosCh=datos.loc[datos['ENTIDAD_NAC']==8]#filtro por estado
datosCh=pd.DataFrame(datosCh)#df de estado

#CONCATENACION
dat2=pd.concat([datosSn,datosNl,datosPb,datosCh],ignore_index= 'False',)#se concatena
dat2h=dat2.loc[dat2['TIPO_PACIENTE']==2]# se filtra por paciente hospitalizado
dat2h=pd.DataFrame(dat2h.groupby(['ENTIDAD_NAC'])['TIPO_PACIENTE'].count())# se agrupa
#ARCHIVO CSV
dat2h.to_csv(r'C:\Users\manuel\tabla2.csv')#se exporta a csv tabla 2


################################################### PARTE 3 ###################################################################


dat2h.plot(kind='bar')#graficar con df directo
plt.savefig('grafica1.png')#guarda imagen


##################################################### PARTE 4 #####################################################################
x=datosc['FECHA']#se descompone uno de los primero datafframes para obtner los dato
y=datosc['CONFIRMADOS']
plt.plot(x,y)#grafica
plt.show()#muestra en pantall
plt.xticks(rotation= 90)# arrgla el acomodo en x
plt.ylabel('CASOS CONFIRMADOS')
plt.xlabel('fecha')
plt.savefig('grafica2.png')#guarda imagen grafica2
```
